# This notebook calculates the total number of ATD participants each fiscal year and puts the data into a plottable format.

```python
#import
import pandas as pd
```

```python
df_19 = pd.read_csv("atd_19.csv")
df_20 = pd.read_csv("atd_20.csv")
df_21 = pd.read_csv("atd_21.csv")
df_22 = pd.read_csv("atd_22.csv")
df_23 = pd.read_csv("atd_23.csv")
df_19
```

```python
df_19["year"] = 2019
df_20["year"] = 2020
df_21["year"] = 2021
df_22["year"] = 2022
df_23["year"] = 2023
```

```python
df_all = pd.concat([df_19, df_20, df_21, df_22, df_23])
df_all
```

```python
df_all.to_csv("allyear_detail.csv", index = False)
```

```python
df_fy = df_all.pivot_table(index = "year", values = "count", aggfunc = "sum").reset_index()
df_fy
```

```python
df_fy.to_csv("total_fy.csv", index = False)
```

```python
df_fy_tech = df_all.pivot_table(index = ["year", "tech"], values = "count", aggfunc = "sum").reset_index()

df_fy_tech
```

```python
df_fy_tech.tech = df_fy_tech.tech.apply(lambda x: "VoiceID" if x == "TR" or x == "No Tech" else x)
df_fy_tech
```

```python
df_fy_tech.to_csv("fy_by_tech.csv", index = False)
```

```python
df_smartlink = df_fy_tech[df_fy_tech.tech == "SmartLINK"]
df_smartlink
```

```python
df_smartlink.to_csv("smartlink-fy.csv", index = False)
```
