### From the section above, it's clear that all the rows without a current or potential end date are IDVs, and are not of concern to this data analysis. This section divivdes all the contract awards by the fiscal years they span across.

````python
# first dropping some more columns that are not useful
df_awards = df[df.award_or_idv_flag == "AWARD"]
df_awards = df_awards.drop(columns = ["current_total_value_of_award", "potential_total_value_of_award", "period_of_performance_potential_end_date", "idv_type", "solicitation_procedures", "last_modified_date"])
df_awards
````

````python
df_awards.info()
````

##### This section calculates the fiscal years from contract start dates and end dates

````python
df_awards["start_date"] = pd.to_datetime(df.period_of_performance_start_date)
````

````python
df_awards["end_date"] = pd.to_datetime(df.period_of_performance_current_end_date)
````

````python
df_awards.info()
````

````python
df_awards["start_fiscal_year"] = df_awards["start_date"].map(lambda x: x.year if x.month < 10 else x.year + 1)
df_awards["end_fiscal_year"] = df_awards["end_date"].map(lambda x: x.year if x.month < 10 else x.year + 1)
df_awards
````

````python
#calculating how many fiscal years an obligated spending spans across

df_awards["years"] = df_awards.end_fiscal_year - df_awards.start_fiscal_year + 1
df_awards
````

````python
# Calculates average obligated amount each fiscal year

df_awards["ave_spending_each_year"] = df_awards.total_obligated_amount/df_awards.years
df_awards.head()
````

````python
# df_sum.apply(lambda row: "MVM INC" if row["recipient_uei"]=="ZHJYA5NN32Z5" else row["recipient"], axis=1)

df_awards["fiscal_years"] = df_awards.apply(lambda row: [df_awards["start_fiscal_year"]].\
                                            append(df_awards["start_fiscal_year"] +1) if row["end_fiscal_year"] > row["start_fiscal_year"])
````
